```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 3
# Create a new set of 2,000 random latitudes and longitudes.
lats=np.random.uniform(low=-90.000,high=90.000, size=2000)
lngs=np.random.uniform(low=-180.000,high=180.000, size=2000)
lat_lngs=zip(lats,lngs)
lat_lngs
```

```python
coordinates = list(lat_lngs)
```

```python
from citipy import citipy
```

```python
# Step 4
# Get the nearest city using the citipy module.
for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
# Create a list for holding the cities.
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    if city not in cities:
        cities.append(city)

```

```python
len(cities)
```

```python
# Import the requests library.
import requests

# Import the API key.
from config import weather_api_key
```

```python
# Step 5
# Perform an API call with the OpenWeatherMap
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Example of city to look at the structure of JSON
city_example='Mexico City'
weather_example=requests.get(url+'&q='+city_example).json()
weather_example
```

```python
city_data_ex=[]
city_lat_ex = weather_example["coord"]["lat"]
city_lng_ex = weather_example["coord"]["lon"]
city_max_temp_ex = weather_example["main"]["temp_max"]
city_humidity_ex = weather_example["main"]["humidity"]
city_clouds_ex = weather_example["clouds"]["all"]
city_wind_ex = weather_example["wind"]["speed"]
city_country_ex = weather_example["sys"]["country"]
city_description_ex=weather_example["weather"][0]["description"]
# Data Frame
city_data_ex.append({"City": city_example.title(),
                     "Lat": city_lat_ex,
                     "Lng": city_lng_ex,
                     "Max Temp": city_max_temp_ex,
                     "Humidity": city_humidity_ex,
                     "Cloudiness": city_clouds_ex,
                     "Wind Speed": city_wind_ex,
                     "Country": city_country_ex,
                     "Description": city_description_ex})
ex_df=pd.DataFrame(city_data_ex)
ex_df
```

```python
# Step 6
city_data = []

for i, city in enumerate(cities):
              
    city_url = url + "&q=" + city.replace(" ","+")

    print(f'City name: {city}')
    
    
        
# Retrieve the following information from the API call:
# 1. Latitude and longitude
# 2. Maximum temperature
# 3. Percent humidity
# 4. Percent cloudiness
# 5. Wind speed
# 6. Weather description (for example, clouds, fog, light rain, clear sky)
    try:
        city_weather = requests.get(city_url).json()
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description=city_weather["weather"][0]["description"]
        # Data Frame
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Description": city_description})

# If an error is experienced, skip the city.
    except:
        print("City not found.")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Finish")
print("-----------------------------")
```

```python
# Step 7
# Add the data to a new DataFrame.
city_weather_df=pd.DataFrame(city_data)
city_weather_df.head()
```

```python
column_order=['City','Country','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed','Description']
city_weather_df=city_weather_df[column_order]
city_weather_df
```

```python
# Step 8
# Export the DataFrame as a CSV file, and save it as WeatherPy_Database.csv in the Weather_Database folder.
save_as = "WeatherPy_Database.csv"
city_weather_df.to_csv(save_as, index_label="City_ID")
```
